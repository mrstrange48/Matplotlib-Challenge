# Winning Wrestlers Entertainment

In this activity you will be taking four seperate csvs that were scraped down from a wrestling database, merging them together, and then creating charts to visualize a wrestler's wins and losses over the course of four years.

### Part 1 - Macho Merging

* You will likely need to perform three different merges over the course of this activity, changing the names of your columns as you go along.

```python
# Import the necessary modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
# Bring each CSV into a separate data frame
WWE2013_df = '../Resources/WWE-Data-2013.csv'
WWE2014_df = '../Resources/WWE-Data-2014.csv'
WWE2015_df = '../Resources/WWE-Data-2015.csv'
WWE2016_df = '../Resources/WWE-Data-2016.csv'

WWE_data_2013_df = pd.read_csv(WWE2013_df)
WWE_data_2014_df = pd.read_csv(WWE2014_df)
WWE_data_2015_df = pd.read_csv(WWE2015_df)
WWE_data_2016_df = pd.read_csv(WWE2016_df)

```

```python
# Merge the first two datasets (2013 and 2014) on "Wrestler" so that no data is lost (should be 182 rows)
combinedf= WWE_data_2013_df.merge(WWE_data_2014_df,on='Wrestler',suffixes=("2013_","2014_"))
combinedf
```

```python
# Rename our _x columns to "2013 Wins", "2013 Losses", and "2013 Draws"
combinedf=combinedf.rename(columns={"Wins2013_":"2013 Wins","Losses2013_":"2013 Losses","Draws2013_":"2013 Draws","Wins2014_"
                                   :"2014 Wins", "Losses2014_":"2014 Losses", "Draws2014_":"2014 Draws"})
combinedf
# Rename our _y columns to "2014 Wins", "2014 Losses", and "2014 Draws"

```

```python
# Merge our newly combined dataframe with the 2015 dataframe
combinedf= combinedf.merge(WWE_data_2015_df,on='Wrestler')
combinedf
```

```python
# Rename "wins", "losses", and "draws" to "2015 Wins", "2015 Losses", and "2015 Draws"
combinedf=combinedf.rename(columns={"Wins":"2015 Wins","Losses":"2015 Losses","Draws":"2015 Draws"})

combinedf
                    
                                
```

```python
# Merge our newly combined dataframe with the 2016 dataframe

combinedf= combinedf.merge(WWE_data_2016_df,on='Wrestler')
combinedf
```

```python
# Rename "wins", "losses", and "draws" to "2016 Wins", "2016 Losses", and "2016 Draws"

combinedf=combinedf.rename(columns={"Wins":"2016 Wins","Losses":"2016 Losses","Draws":"2016 Draws"})
combinedf
```
